```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
formula ='average_price~ a+b+c+d+e+f+g+h+i'
```

```python
anova_results = anova_lm(ols(formula,data).fit())
print(anova_results)
```

```python
formula1 ='average_price~a*b+a*g+b*g'
```

```python
anova_results1 = anova_lm(ols(formula1,data).fit())
print(anova_results1)
```

```python
from sklearn.cluster import KMeans
```

```python
x= pd.DataFrame(data,columns = ['average_price','a','b','g'])
x
```

```python
kms = KMeans(n_clusters = 3)
y = kms.fit_predict(x)
print(y)
```

```python
mask1 = y ==0
mask2 = y ==1
mask3 = y ==2
mask4 = y ==3
```

```python
plt.figure()
plt.scatter(x['average_price'][mask1],x['a'][mask1],x['b'][mask1],x['g'][mask1])
plt.scatter(x['average_price'][mask2],x['a'][mask2],x['b'][mask2],x['g'][mask2])
plt.scatter(x['average_price'][mask3],x['a'][mask3],x['b'][mask3],x['g'][mask3])
plt.scatter(x['average_price'][mask4],x['a'][mask4],x['b'][mask4],x['g'][mask4])
plt.show()
```
